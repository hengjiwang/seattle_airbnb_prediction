```python
def num_att(pos,attr):
    dis = attr.apply(lambda x: distance.distance(x[1:],pos).miles,axis=1)
    return (dis<1).sum()
```

```python
myl = myl.iloc[0:100]
```

```python
myl['num'] = myl.apply(lambda x: num_att(x,sea_att),axis=1)
```

```python
myl['w']=0
myl['p']=0
```

```python
myl.loc[myl['num']==1,['w','p']]=[2,3]
```
